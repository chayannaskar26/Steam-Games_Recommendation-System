### Fixing string in Supported Language column

```python
import re
reg = re.compile('<.*?>') 

def cleanhtml(text):
  final_text = re.sub(reg, '', text)
  return final_text
```

```python
steam_df['supported_languages'] = steam_df['supported_languages'].apply(lambda x: cleanhtml(x))
```

```python
steam_df.head()
```

```python
steam_df['platforms'][0]
```

```python
import ast

def getPlatforms(obj):
    list = []
    for i in ast.literal_eval(obj).items():
        if(i[1] == True):
            list.append(i[0])
    return list
```

```python
steam_df['platforms'] = steam_df['platforms'].apply(getPlatforms)
```

```python
steam_df.head()
```

```python
steam_df['categories'][0]
```

```python
import ast

def getListValue(obj):
    list = []
    for i in ast.literal_eval(obj):
        for key, value in i.items():
            if(key == 'description'):
                list.append(value)
                
    return list
            
```

```python
steam_df['categories'] = steam_df['categories'].apply(getListValue)
steam_df['genres'] = steam_df['genres'].apply(getListValue)
```

```python
steam_df.head()
```

```python
free_paid_dict = {True: 'Free', False: 'Paid'}
```

```python
steam_df = steam_df.replace({'is_free': free_paid_dict})
```

```python
steam_df['short_description'] = steam_df['short_description'].apply(lambda x: x.split())
```

```python
steam_df['categories'] = steam_df['categories'].apply(lambda x: [i.replace(' ', '') for i in x])
steam_df['genres'] = steam_df['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
steam_df.head()
```

```python
steam_df['developers'][0]
```

```python
steam_df['developers'] = steam_df['developers'].apply(lambda x: [i.replace(' ', '') for i in ast.literal_eval(x)])
steam_df['publishers'] = steam_df['publishers'].apply(lambda x: [i.replace(' ', '') for i in ast.literal_eval(x)])
```

```python
steam_df['release_year'] = steam_df['release_date'].apply(lambda x: x.split(' ')[-1])
```

```python
steam_df.head()
```

```python
steam_df['is_free'] = steam_df['is_free'].apply(lambda x: [x])
steam_df['release_year'] = steam_df['release_year'].apply(lambda x: [x])
steam_df['supported_languages'] = steam_df['supported_languages'].apply(lambda x: x.split(', '))
```

```python
steam_df.head()
```

```python
steam_df['tags'] = steam_df['short_description'] + steam_df['is_free'] + steam_df['supported_languages'] + steam_df['developers'] + steam_df['publishers'] + steam_df['categories'] + steam_df['genres'] + steam_df['release_year']
```

```python
steam_df.head()
```

```python
new_df = steam_df[['appid', 'title', 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stemText(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return ' '.join(y)
```

```python
new_df['tags'].apply(stemText)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommendGame(gameID):
    index = new_df[new_df['appid'] == gameID].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommendGame(242050)
```
